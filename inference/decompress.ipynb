# chunking and padding

```python
####### compressed chunks
import os
import hashlib

block_size = 512 * 1024  # 512 KiB

def pad_block(block):
    """padding to 512 KiB"""
    if len(block) < block_size:
        return block + b'\x00' * (block_size - len(block))
    return block

def process_file(input_file, output_directory):
    with open(input_file, 'rb') as infile:
        block_num = 0
        while True:
            block = infile.read(block_size)
            if not block:
                break
            block = pad_block(block)
            hash_digest = hashlib.md5(os.path.abspath(input_file).encode()).hexdigest()
            output_file = os.path.join(output_directory, f"{hash_digest}_{block_num}.bin")
            with open(output_file, 'wb') as outfile:
                outfile.write(block)
            block_num += 1

def find_files(directory):
    for root, _, files in os.walk(directory):
        for file in files:
            yield os.path.join(root, file)

directory_to_search = './compressed_data_ubuntu/'
output_directory = './compressed_data_ubuntu_512KiB/'
os.makedirs(output_directory, exist_ok=True) 
all_files = list(find_files(directory_to_search))

cnt = 0
for file in all_files:
    process_file(file, output_directory)
    cnt += 1
    print(cnt, end='\r')
```
